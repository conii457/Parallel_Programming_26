```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Columnas:
# 0 -> iteracion
# 1 -> magnetizacion total
# 2 -> magnetizacion por sitio
datos = np.loadtxt("ising_1.dat", comments="#")

i = datos[:, 0]
mtot = datos[:, 1]/256
msitio = datos[:, 2]

plt.plot(i, mtot, linewidth=2, label="T = 2.0")

plt.title("Modelo de Ising: Magnetización total vs Iteraciones")
plt.xlabel("Iteraciones")
plt.ylabel("Magnetización total")
plt.grid()
plt.legend()
plt.show()
```

```python
# columnas: T, r, C(r)
datos = np.loadtxt("ising_2_corr.dat")

T = datos[:, 0]
r = datos[:, 1]
C = datos[:, 2]
temperaturas = [1.5, 2.2, 2.3, 3.5]

for T_obj in temperaturas:
    idx = np.isclose(T, T_obj, atol=1e-6)
    plt.plot(
        r[idx],
        C[idx],
        linewidth=1.5,
        label=f"T = {T_obj:.1f}"
    )

plt.xlabel("r")
plt.ylabel("C(r)")
plt.title("Función de correlación para distintas temperaturas")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# columnas = T, xi
datos = np.loadtxt("ising_2_xi.dat")

T = datos[:, 0]
xi = datos[:, 1]
plt.plot(T, xi, marker='o', linewidth=1.5)

plt.xlabel("Temperatura")
plt.ylabel(r'$\xi$')
plt.title("Longitud de correlación en función de la temperatura")
plt.grid(True)
plt.tight_layout()
plt.show()
```
